```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['font.sans-serif'] = ['WenQuanYi Zen Hei', 'SimHei', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False
# -------------------------- 1. 准备实验数据 --------------------------
E = np.array([0.50, 1.01, 1.49, 2.00, 2.49, 3.00, 3.50])
S = np.array([0.0034, 0.0072, 0.0103, 0.0141, 0.0172, 0.0210, 0.0243])


# -------------------------- 2. 线性拟合（最小二乘法） --------------------------
# np.polyfit(自变量, 因变量, 拟合多项式阶数) ，这里拟合一阶直线（S = a*E + b）
coefficients = np.polyfit(E, S, 1)
a, b = coefficients  # a 是斜率，b 是截距
print(f"拟合得到的直线方程为：S = {a:.4f}E + {b:.4f}")  # 打印拟合方程（可选）


# -------------------------- 3. 生成拟合曲线的 x 轴数据 --------------------------
# 在 E 的取值范围内生成 100 个均匀分布的点，用于绘制平滑的拟合曲线
x_fit = np.linspace(min(E), max(E), 100)
y_fit = a * x_fit + b  # 代入拟合方程计算对应的 y 值


# -------------------------- 4. 绘制图表 --------------------------
plt.figure(figsize=(8, 6))  # 设置画布大小（宽8英寸，高6英寸）

# 绘制实验数据点（黑色散点）
plt.scatter(E, S, color='black', label='实验数据点')

# 绘制拟合曲线（蓝色实线）
plt.plot(x_fit, y_fit, 'b-', label=f'拟合曲线: S = {a:.4f}E + {b:.4f}')

# 设置坐标轴标签、标题、范围、网格、图例
plt.xlabel('E (V)')          # x 轴标签
plt.ylabel('S')              # y 轴标签
plt.title('E - S 图')        # 图表标题
plt.xlim(0, 4)               # x 轴范围（与示例一致）
plt.ylim(0, 0.03)            # y 轴范围（与示例一致）
plt.grid(True)               # 显示网格线
plt.legend()                 # 显示图例
plt.savefig('E-S图.png', dpi=300, bbox_inches='tight')  # 保存为高分辨率 PNG 图片
plt.show()  # 显示图表
```
